### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x**2

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2', color='blue')

# Customizations
plt.title('Line Plot of y = x²')
plt.xlabel('x values')
plt.ylabel('y values (x squared)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
