### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']  # Different colors for each bar

# Create the bar plot
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=colors)

# Add title and labels
plt.title('Bar Plot of Category Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```
